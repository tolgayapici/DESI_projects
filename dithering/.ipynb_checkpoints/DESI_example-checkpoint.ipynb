```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import glob
import os
from astropy.io import fits
from astropy import units
from astropy import wcs
```

```python
list_loss = ['star', 'lrg', 'elg']
for loss_source in list_loss:
    fiberlossfile = os.environ["DESIMODEL"]+"/data/throughput/fiberloss-%s.dat"%loss_source
    x, y = np.loadtxt(fiberlossfile, unpack=True)
    plt.plot(x, y, label="%s"%loss_source)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Fiber Acceptance")
plt.show()
```

```python
blur = os.environ["DESIMODEL"]+"/data/throughput/DESI-0347_blur.ecsv"
data = np.genfromtxt(blur, unpack=True, skip_header=35)
allfile = open(blur).readlines()
labels = allfile[34].split('"')
for i in range(len(data)-1):
    plt.plot(data[0], data[1+i], label=labels[1+i*2])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Offset [$\mu$ m]")
plt.show()
```

```python
offset = os.environ["DESIMODEL"]+"/data/throughput/DESI-0347_offset.ecsv"
data = np.genfromtxt(offset, unpack=True, skip_header=36)
allfile = open(offset).readlines()
labels = allfile[35].split('"')
for i in range(len(data)-1):
    plt.plot(data[0], data[1+i], label=labels[1+i*2])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.xlabel("Wavelength [$\AA$]")
plt.ylabel("Centroid Offset [$\mu$ m]")
plt.show()
```

```python
def get_r_random_offset(seed):
    roffset = os.environ["DESIMODEL"]+"/data/throughput/DESI-0347_random_offset_1.fits"
    data = fits.open(roffset)
    offsetx = data[1].data
    offsety = data[2].data
    offsetr = np.sqrt(offsetx ** 2 + offsety ** 2)
    dx = data[1].header['CDELT1']
    dy = data[2].header['CDELT1']
    num_x = len(offsetx)
    w = wcs.WCS(data[1].header)
    xlo, ylo, xhi, yhi = w.wcs_pix2world([[0, 0], [num_x-1, num_x-1]], 0).flatten()
    coord = np.linspace(xlo, xhi, num_x)
    img = plt.quiver(coord, coord, offsetx, offsety, offsetr)
    plt.clf()
    plt.colorbar(img)
    plt.tight_layout()
    plt.savefig("random_offset_r_seed%d.png"%seed)
    return coord, offsetx, offsety, offsetr

c, dx, dy, dr = get_r_random_offset(2)
```
